二次関数の場合、微分することで最適解(いわゆるロスが0)は一発で求まる。
しかしパラメータの多い機械学習では難しい。
->よって勾配降下法を用いる。

### 勾配降下法

xi+1=xi − ηd/dx f(x)　　　　η…学習率
で更新していく

```python
def f(x):
    return x**2 - 4*x + 5

def df(x):
    return 2*x - 4

#接戦
def tangent(x, xx):
    a = df(xx)
    b = f(xx) - a*xx
    return a*x + b

#xxの更新
def gradient_descent(xx, eta=0.1, n_iter=3):
    path=[]
    for _ in range(n_iter):
        #傾き
        grad = df(xx)
        path.append(grad)
        xx = xx-eta*grad
    return path
```

```python
x = np.linspace(0,4,40)
xx = 4
path_list = gradient_descent(xx) #[4, 3.2, 2.5600000000000005]
```

```python
fig, axes = plt.subplots(1,3,figsize=(14,4))
plt.subplots_adjust(wspace=0.3)
for i in range(3):
    axes[i].set_title('n = {}'.format(i+1))
    axes[i].set_xlim([0, 4])
    axes[i].set_ylim([0.5, 5])
    axes[i].plot(x, f(x))
    axes[i].plot(x, tangent(x, path_list[i]))   
```

### 最急降下法アルゴリズム

```python
class GradientDescent:
    def __init__(self, f, df, eta=0.05, eps=1e-6):
        self.f = f
        self.df = df
        self.eta = eta
        self.eps = eps
        self.path = None
    
    def solve(self, x):
        path = []
        grad = self.df(x)
        path.append(x)
        
        #多変数の場合のノルムの大きさとself.epsを比較
        #self.epsの２乗より大きい場合に更新を続ける
        while(grad**2).sum() > self.eps**2:
            x = x - self.eta * grad
            grad = self.df(x)
            path.append(x)
        self.path_ = np.array(path)
        self.x_ = x
        self.opt_ = f(x)
        self.grad_ = grad
```

実際に**f(x,y)=6x2−xy+y2+x+2yf(x,y)=6x2−xy+y2+x+2y**で計算する

```python
def f(xy):
    x,y = xy[0],xy[1]
    return 6*x**2 - x*y + y**2 + x + 2*y

def df(xy):
    x,y = xy[0],xy[1]
    return np.array([12*x - y + 1, - x + 2*y + 2])
```

```python
gd = GradientDescent(f,df)
initial = np.array([1,1])
gd.solve(initial)
```

```python
np.set_printoptions(precision=7, suppress=True)    # 有効桁数の設定
print('更新後の変数: {}'.format(gd.x_))    # 更新後の変数: [-0.173913 -1.086956]
print('更新後のf(x,y): {:.3}'.format(gd.opt_))    # 更新後のf(x,y): -1.17
print('更新後の勾配: {}'.format(gd.grad_)) 
```

```python
plt.plot(gd.path_[:, 0], gd.path_[:, 1]);
```
