```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r = np.arange(0, 2, 0.01)
theta = 6*np.pi*r
```

```python
k = 100
x = r*np.cos(theta)*np.exp(theta/k)
y = r * np.sin(theta) * np.exp(theta / k)
```

```python

plt.scatter(x, y)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()
```

```python
A = np.vstack([x, y]).T
A.shape
```

ReLU関数

```python
def relu(x):
    return np.maximum(0,x)
```

```python
def inv_manifold(A, n_dim):
    """
    A: [200,2]
    T: [2,n_dim]
    """
    #[2xndim]の行列を作成
    T = np.random.randn(2,n_dim)
  
    #逆行列に変換
    pinvT = np.linalg.pinv(T)

    invA = relu(A@T)@pinvT
    return invA
```

```python
plt.figure(figsize=(20,5))
for i, n_dim in enumerate([2,3,5,15,30]):
    invA = inv_manifold(A, n_dim)
    plt.subplot(1, 5, i+1)
    plt.scatter(invA[:, 0], invA[:, 1])
    plt.title('Output/dim='+str(n_dim))
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
plt.show()
```
